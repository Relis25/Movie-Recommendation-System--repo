```python
#String of list convert into list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_cleaned['genres'].apply(convert)
```

```python
movies_cleaned = movies_cleaned.copy()
movies_cleaned['genres']=movies_cleaned['genres'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies_cleaned['cast'].apply(convert3)
```

```python
movies_cleaned = movies_cleaned.copy()
movies_cleaned['cast']=movies_cleaned['cast'].apply(convert3)
```

```python
movies_raw['crew'][0]       # -- >from this line use only "name" dictionary when "job": "Director"

#--- > {"credit_id": "52fe48009251416c750ac9c3", "department": "Directing", "gender": 2, "id": 2710, "job": "Director", "name": "James Cameron"}, 
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L      
```

```python
movies_raw['crew'].apply(fetch_director)
```

```python
movies_cleaned = movies_cleaned.copy()
movies_cleaned['crew']=movies_raw['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)   
```

```python
# Use for movie recommend function(2)

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]     #----> give value from index 1 to 5
```

```python
def recommend(movie):
    index = movies_structured[movies_structured['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(movies_structured.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
